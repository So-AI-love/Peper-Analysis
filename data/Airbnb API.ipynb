```python
import airbnb
api = airbnb.Api('martin_datsev00@abv.bg', 'pesho123')
```

```python
sofia = api.get_homes_with_query('Sofia')
```

```python
import json
```

```python
items_per_grid = 100
for i in range(5, 10):
    resp = api.get_homes_with_query('Sofia', offset=i*items_per_grid, items_per_grid=items_per_grid)
    f = open('sofia-%d-%d.json' % (i*items_per_grid, (i+1)*items_per_grid - 1), "w+")
    f.write(json.dumps(resp))
    f.close()
    if not resp["explore_tabs"][0]["pagination_metadata"]["has_next_page"]:
        break
```

```python
ids = []
listings = []
items_per_grid = 100
for i in range(0, 5):
    f = open('sofia-%d-%d.json' % (i*items_per_grid, (i+1)*items_per_grid - 1), "r")
    resp = json.loads(f.read())
    f.close()
    ids += resp["explore_tabs"][0]["home_tab_metadata"]["remarketing_ids"]
    listings += resp["explore_tabs"][0]["sections"][0]["listings"]
ids = list(set(ids))
ids
```

```python
all_listings = []
for id in ids:
    for l in listings:
        if l["listing"]["id"] == id:
            all_listings.append(l)
            break
```

```python
import csv
```

```python
limit = 100
counter = 0
f = open('all_listings.ndjson', 'w+')
for l in all_listings:
    id = l["listing"]["id"]
    reviews = api.get_reviews(id, offset=0, limit=limit)
    count = reviews["metadata"]["reviews_count"]
    reviews = reviews["reviews"]
    print(counter, ' count ', count)
    counter += 1
    if count > limit:
        for i in range(count // limit):
            reviews += api.get_reviews(id, offset=(i+1)*limit, limit=limit)["reviews"]
            print(len(reviews))
    l["reviews"] = reviews
    f.write(json.dumps(l) + '\n')
```

```python
c = 0
for l in all_listings:
    c += len(l["reviews"])
c
```

```python
f = open('all_reviews.csv', 'wb')
csvwriter = csv.writer(f)
csvwriter.writerow(['lat', 'lng', 'reviews'])
with open('all_listings.ndjson') as fp:
    for line in fp:
        l = json.loads(line)
        lat = l["listing"]["lat"]
        lng = l["listing"]["lng"]
        reviews = map(lambda r: r["comments"], l["reviews"])
        csvwriter.writerow([lat, lng, json.dumps(reviews)])
f.close()
```
